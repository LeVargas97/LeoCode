```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sinistros_seguros.csv', sep=',')
```

```python
print(df)
```

```python


print(df.head(1))
```

```python

print(df.isnull().sum())
```

```python
df = df.dropna()


print(df)
```

```python

df['Data_Sinistro'] = pd.to_datetime(df['Data_Sinistro'])
print(df)
```

```python
df['Mes_Sinistro'] = df['Data_Sinistro'].dt.month
print(df)
```

```python

df['Tipo_seguro'] = df['Produto_Seguro'].apply(lambda x: "Automovel" if 'auto' in x.lower() else("Residencial" if 'residencial' in x.lower() else "vida"))
print(df)
```

```python
# prompt: faça resumo estatistico basico

print(df.describe())
```

```python
# NUMERO TOTAL DE SINISTROS
sinistro_por_tipo = df.groupby('Tipo_seguro')["Valor_Indenizacao"].agg(['count', 'sum'])

#ESTATISTICA POR VALOR DAS INDENIZAÇÕES
estatisticas_indenizacao = df['Valor_Indenizacao'].describe()

#IDENTIFICAR USANDO IQR
Q1 = df['Valor_Indenizacao'].quantile(0.25)
Q3 = df['Valor_Indenizacao'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Valor_Indenizacao'] < Q1 - 1.5 * IQR) | (df['Valor_Indenizacao'] > Q3 + 1.5 * IQR)]



print(f"sinistro por tipo: \n{sinistro_por_tipo}")
print('#################################################')

print(f"estatisticas por indenizacao: \n{estatisticas_indenizacao}")

print('#################################################')


print(f"Outliers: \\n{outliers}")
```

```python
sinistros_mensais = df.groupby(['Mes_Sinistro', 'Tipo_seguro'])['Valor_Indenizacao'].count().unstack()


print(sinistros_mensais)
```

```python
# GRAFICO DISTRIBUICAO VALORES DAS INDENIZAÇÕES

plt.figure(figsize=(10, 6))

df['Valor_Indenizacao'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title('Distribuição dos Valores das Indenizações')
plt.xlabel('Valores das Indenizações(R$)')
plt.ylabel('Frequência')
plt.grid(True)
```

```python
# faça grafico de frequencias de sinistros mensais por tipo de seguro


plt.figure(figsize=(12, 6))
sinistros_mensais.plot(kind='line', marker='o', linestyle='-', colormap='viridis')
plt.title('Sinistros Mensais por Tipo de Seguro')
plt.xlabel('Mês')
plt.ylabel('Número de Sinistros')
plt.grid(True)
plt.show()
```

```python
#grafico valor de indenização vc quantidade de sinistro

plt.figure(figsize=(10, 6))
plt.scatter(df['Quantidade_Sinistros'], df['Valor_Indenizacao'],alpha=0.5, color='purple')
plt.title('Relação entre Quantidade de sinistris e valor de indenização')
plt.xlabel('Quantidade de Sinistros')
plt.ylabel('Valor de Indenização(R$)')
plt.grid(True)
plt.show()
```
