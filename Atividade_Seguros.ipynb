```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python

df_sinistros = pd.read_csv('sinistros_seguros.csv')


df_sinistros.head()
```

```python
df_sinistros['Data_Sinistro'] = pd.to_datetime(df_sinistros['Data_Sinistro'])


print(df_sinistros)
```

```python
df_sinistros['Mes_Sinistro'] = df_sinistros['Data_Sinistro'].dt.month


print(df_sinistros.head())
```

```python
sinistros_por_tipo = df_sinistros.groupby('Produto_Seguro').agg({'Valor_Indenizacao': 'sum', 'Quantidade_Sinistros': 'count'})


sinistros_por_tipo = sinistros_por_tipo.rename(columns={'Valor_Indenizacao': 'Valor_Total_Indenizacoes', 'Numero_Sinistro': 'Total_Sinistros'})

print(sinistros_por_tipo)
```

```python
estatisticas_indenizacoes = df_sinistros['Valor_Indenizacao'].describe()


print(estatisticas_indenizacoes)


media_indenizacoes = df_sinistros['Valor_Indenizacao'].mean()
print(f"Média das indenizações: {media_indenizacoes}")


mediana_indenizacoes = df_sinistros['Valor_Indenizacao'].median()
print(f"Mediana das indenizações: {mediana_indenizacoes}")


desvio_padrao_indenizacoes = df_sinistros['Valor_Indenizacao'].std()
print(f"Desvio padrão das indenizações: {desvio_padrao_indenizacoes}")
```

```python
Q1 = df_sinistros['Valor_Indenizacao'].quantile(0.25)
Q3 = df_sinistros['Valor_Indenizacao'].quantile(0.75)
IQR = Q3 - Q1


limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR


outliers = df_sinistros[(df_sinistros['Valor_Indenizacao'] < (Q1 - 1.5 * IQR)) | (df_sinistros['Valor_Indenizacao'] > (Q3 + 1.5 * IQR))]


print("Possíveis outliers nos valores das indenizações:", outliers)
```

```python
evolucao_sinistros = df_sinistros.groupby(['Produto_Seguro', 'Mes_Sinistro'])['Quantidade_Sinistros'].sum().unstack()

print(evolucao_sinistros)


```

```python
evolucao_sinistros.plot(kind='line', figsize=(10, 6))
plt.title('Evolução do Número de Sinistros por Tipo de Seguro')
plt.xlabel('Mês do Sinistro')
plt.ylabel('Número de Sinistros')
plt.legend(title='Tipo de Seguro')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df_sinistros['Valor_Indenizacao'], bins=20)
plt.title('Distribuição dos Valores das Indenizações')
plt.xlabel('Valor da Indenização')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df_sinistros['Valor_Indenizacao'])
plt.title('Boxplot dos Valores das Indenizações')
plt.ylabel('Valor da Indenização')
plt.show()
```

```python
sinistros_por_tipo_mes = df_sinistros.groupby(['Produto_Seguro', 'Mes_Sinistro'])['Quantidade_Sinistros'].sum().unstack()


for tipo_seguro in sinistros_por_tipo_mes.index:
  plt.figure(figsize=(10, 6))
  plt.plot(sinistros_por_tipo_mes.loc[tipo_seguro])
  plt.title(f'Frequência de Sinistros para {tipo_seguro}')
  plt.xlabel('Mês')
  plt.ylabel('Número de Sinistros')
  plt.xticks(range(1, 13))
  plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_sinistros['Valor_Indenizacao'], df_sinistros['Quantidade_Sinistros'])
plt.title('Relação entre Valor da Indenização e Quantidade de Sinistros')
plt.xlabel('Valor da Indenização')
plt.ylabel('Quantidade de Sinistros')
plt.show()
```
